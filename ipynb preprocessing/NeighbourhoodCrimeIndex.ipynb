```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
```

```python
crime_data = pd.read_excel('neighbourhood-crime-rates.xlsx').drop(['_id', 'OBJECTID', 'geometry'], axis=1)
crime_data
```

```python
labels_2020 = ['Neighbourhood','Hood_ID','Assault_2020', 'AutoTheft_2020', 'BreakAndEnter_2020', 'Robbery_2020',
              'TheftOver_2020', 'Homicide_2020', 'Shootings_2020']
crimes = ['Assault_2020', 'AutoTheft_2020', 'BreakAndEnter_2020', 'Robbery_2020',
              'TheftOver_2020', 'Homicide_2020', 'Shootings_2020']
data_2020 = crime_data[labels_2020].copy()
scaler = MinMaxScaler(feature_range=(0,10))
data_2020[crimes] = scaler.fit_transform(data_2020[crimes])
data_2020[crimes] = data_2020[crimes]
data_2020['crimeIndex'] = data_2020[crimes].mean(axis=1)
data_2020
```

```python
data_2020.crimeIndex.describe()
```

```python
mean = data_2020.crimeIndex.mean()
quantile_25 = data_2020.crimeIndex.quantile(0.25)
quantile_50 = data_2020.crimeIndex.quantile(0.5)
quantile_75 = data_2020.crimeIndex.quantile(0.75)

print(f"Mean = {mean} \nlow crime threshold = >{quantile_25} \nmedium crime threshold = >{quantile_50}\nhigh crime threshold = >{quantile_75}")
```

```python
data_2020.to_csv('Neighbourhood_crimeIndex.csv')
```
