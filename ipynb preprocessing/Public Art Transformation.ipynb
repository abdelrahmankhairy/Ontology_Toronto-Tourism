```python
from geopy.geocoders import Nominatim 
import pandas as pd 
import numpy as np
import copy
```

```python
art_public = pd.read_excel('Public_Art.xlsx')
art_public.head()
```

```python
# locationdf = art_public[['_id','Location']]
# locationdf.head()  
locationdf = copy.deepcopy(art_public) 

#remove those with multiple address or intersection 
locationdf = locationdf[locationdf.Location.str.contains("and") == False]
locationdf = locationdf.reset_index(drop=True)
locationdf.head()
```

```python
#Seperate the full address with the building name if it exsit and store in different column 
new_ldf = pd.concat([locationdf, locationdf['Location'].str.split(', ', expand=True)], axis=1)
new_ldf = new_ldf.rename({0: 'a1', 1: 'a2', 2:'a3'}, axis=1)  #rename column 
new_ldf.head()
```

```python
# add city name to location to ensure avoid retrieving location of different place with the same name 
new_ldf['a1'] = new_ldf['a1'].astype(str) + ' Toronto' 

for i in range(len(new_ldf)):
    if new_ldf.iloc[i,29] != None:
        new_ldf.iloc[i,29] = new_ldf.iloc[i,29] + ' Toronto'
        
for i in range(len(new_ldf)):
    if new_ldf.iloc[i,30] != None:
        new_ldf.iloc[i,30] = new_ldf.iloc[i,30] + ' Toronto'
```

```python
# use geopy and try to get a postal for the address 
geolocator = Nominatim(user_agent = "geoapiExercises")
fsr_list = []
ldu_list = []
full_address = [] 
for i in range(len(new_ldf)): 
    try: 
        place = new_ldf.iloc[i,28]
        location = geolocator.geocode(place)
        data = location.raw
        loc_data = data['display_name'].split()
        if len(loc_data[-3]) != 3:
            fsr_list.append(loc_data[-2])
            ldu_list.append(loc_data[-1])
        else:
            fsr_list.append(loc_data[-3])
            ldu_list.append(loc_data[-2])
        full_address.append(data.get('display_name'))
    except:
        fsr_list.append(np.nan)
        ldu_list.append(np.nan)
        full_address.append(np.nan)
geolocator = Nominatim(user_agent = "geoapiExercises")
fsr_list2 = []
ldu_list2 = []
full_address2 = [] 
for i in range(len(new_ldf)): 
    try: 
        place = new_ldf.iloc[i,29]
        location = geolocator.geocode(place)
        data = location.raw
        loc_data = data['display_name'].split()
        if len(loc_data[-3]) != 3:
            fsr_list2.append(loc_data[-2])
            ldu_list2.append(loc_data[-1])
        else:
            fsr_list2.append(loc_data[-3])
            ldu_list2.append(loc_data[-2])
        full_address2.append(data.get('display_name'))
        
    except:
        fsr_list2.append(np.nan)
        ldu_list2.append(np.nan)
        full_address2.append(np.nan)
geolocator = Nominatim(user_agent = "geoapiExercises")
fsr_list3 = []
ldu_list3 = []
full_address3 = [] 
for i in range(len(new_ldf)): 
    try: 
        place = new_ldf.iloc[i,30]
        location = geolocator.geocode(place)
        data = location.raw
        loc_data = data['display_name'].split()
        if len(loc_data[-3]) != 3:
            fsr_list3.append(loc_data[-2])
            ldu_list3.append(loc_data[-1])
        else:
            fsr_list3.append(loc_data[-3])
            ldu_list3.append(loc_data[-2])
        full_address3.append(data.get('display_name'))
    except:
        fsr_list3.append(np.nan)
        ldu_list3.append(np.nan)
        full_address3.append(np.nan)
```

```python
#for address with the value None it will return Piemonte, so we replace them with nan
#For some large places like park, geopy does not return a postal code and we get Ontario instead, so we remove those as well 
fsr_list = [np.nan if x == 'Piemonte,' or x == 'Ontario,' else x for x in fsr_list]
fsr_list2 = [np.nan if x == 'Piemonte,' or x == 'Ontario,' else x for x in fsr_list2]
fsr_list3 = [np.nan if x == 'Piemonte,' or x == 'Ontario,'else x for x in fsr_list3]
#add these fsr to the original dataframe 
new_ldf['fsr1']=fsr_list
new_ldf['fsr2']=fsr_list2
new_ldf['fsr3']=fsr_list3
new_ldf.head()
```

```python
len(ldu_list2)
```

```python
#for address with the value None it will return Piemonte, so we replace them with nan
#For some large places like park, geopy does not return a postal code and we get Ontario instead, so we remove those as well 
ldu_list = [np.nan if x == 'Canada' or x == 'Italia' else x for x in ldu_list]
ldu_list2 = [np.nan if x == 'Canada' or x == 'Italia' else x for x in ldu_list2]
ldu_list3 = [np.nan if x == 'Canada' or x == 'Italia'else x for x in ldu_list3]
#add these fsr to the original dataframe 
new_ldf['ldu1']=ldu_list
new_ldf['ldu2']=ldu_list2
new_ldf['ldu3']=ldu_list3
new_ldf.head()
```

```python
'None, Torino, Piemonte, Italia'
#for address with the value None it will return Piemonte, so we replace them with nan
#For some large places like park, geopy does not return a postal code and we get Ontario instead, so we remove those as well 
full_address = [np.nan if x == 'None, Torino, Piemonte, Italia' else x for x in full_address ]
full_address2 = [np.nan if x == 'None, Torino, Piemonte, Italia' else x for x in full_address2]
full_address3 = [np.nan if x == 'None, Torino, Piemonte, Italia' else x for x in full_address3]
#add these fsr to the original dataframe 
new_ldf['full_address1'] = full_address
new_ldf['full_address2']= full_address2
new_ldf['full_address3']= full_address3
new_ldf.head()
```

```python
#if we have nan value for all three fsr columns that means, we were not able to retrieve a fsr from geopy then in that case, we will drop these rows 
new_ldf = new_ldf.dropna(subset=['fsr1', 'fsr2','fsr3'], how='all')
new_ldf = new_ldf.dropna(subset=['full_address1', 'full_address2','full_address3'], how='all')
new_ldf = new_ldf.dropna(subset=['ldu1', 'ldu2','ldu3'], how='all')
new_ldf = new_ldf.reset_index(drop=True)
ldf2 = copy.deepcopy(new_ldf)
ldf2['FSR'] = new_ldf['fsr1']
ldf2['LDU'] = new_ldf['ldu1']
ldf2['Full_Address'] = new_ldf['full_address1']
ldf2
```

```python
#we are left with locations that contain the fsr with at least one of the fsr column
#we will combine the result
for i in range(len(ldf2)):
    if ldf2.iloc[i,40] == np.nan:
        ldf2.iloc[i,40] = new_ldf.iloc[i,32]
    
#     if ldf2.iloc[i,34] == np.nan:
#         ldf2.iloc[i,34] = new_ldf.iloc[i,33]

for i in range(len(ldf2)):
    if ldf2.iloc[i,41] == np.nan:
        ldf2.iloc[i,41] = new_ldf.iloc[i,35]
        
for i in range(len(ldf2)):
    if ldf2.iloc[i,42] == np.nan:
        ldf2.iloc[i,42] = new_ldf.iloc[i,38]
```

```python
still_null_fsr = ldf2[ldf2['FSR'].isnull()]
null_idx_fsr = still_null_fsr.index.tolist()
still_null_fsr
```

```python
still_null_ldu = ldf2[ldf2['LDU'].isnull()]
null_idx_ldu = still_null_ldu.index.tolist()
still_null_ldu
```

```python
still_null_fa = ldf2[ldf2['Full_Address'].isnull()]
null_idx_fa = still_null_fa.index.tolist()
still_null_fa
```

```python
for i in null_idx_fsr:
    ldf2.iloc[i,40] = ldf2.iloc[i,32]
ldf2['FSR'].isnull().values.any()
```

```python
for i in null_idx_ldu:
    ldf2.iloc[i,41] = ldf2.iloc[i,35]
ldf2['LDU'].isnull().values.any()
```

```python
for i in null_idx_fa:
    ldf2.iloc[i,42] = ldf2.iloc[i,38]
ldf2['Full_Address'].isnull().values.any()
```

```python
#some fsr includes comma, remove the commas 
ldf2['FSR'] = ldf2['FSR'].str.replace(',','')
ldf2 = ldf2.drop(['a1', 'a2', 'a3', 'fsr1', 'fsr2', 'fsr3','ldu1','ldu2', 'ldu3', 'full_address1', 'full_address2', 'full_address3'], axis = 1)
ldf2
```

```python
#save csv 
ldf2.to_csv("Public_Art_Cleaned.csv", index = False)
```
